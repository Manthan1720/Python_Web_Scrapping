```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL="https://www.vijaysales.com/bucketdetails/vs-exclusive/3391"
```

```python
headers=({"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36"})
```

```python
page=requests.get(URL,headers=headers)
page
```

```python
type(page.content)
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
BrandName=[]
```

```python
wn=soup.find_all("h2",class_="BcktPrdNm_")
```

```python
len(wn)
```

```python
for i in wn:
    BrandName.append(i.text)
print(BrandName)
```

```python
TVPrices=[]
```

```python
wp=soup.find_all("div",class_="vspmn_ h21")
```

```python
for i in wp:
    TVPrices.append(i.text)
print(TVPrices)
```

```python
len(TVPrices)
```

```python
Discount=[]
```

```python
wm=soup.find_all("span",class_="ofrnm_")
```

```python
wm
```

```python
for i in wm:
    Discount.append(i.text)
print(Discount)
```

```python
data={"Television":BrandName,"Price":TVPrices,"Discount":Discount}
```

```python
df=pd.DataFrame(data)
```
